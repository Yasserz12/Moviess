import necessary libraries


```python
import pandas as pd
```

```python
# load the data
df = pd.read_excel("movies.xlsx")
```

```python
# explore our data 
df.head()
```

we can notice that the columns names are not consistent so we will fix that 

```python
# make all the columns in lower case 
df.columns = [x.lower() for x in df.columns]
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
# checking for nulls 
df.isnull().sum()
```

#### we will drop the columns ("title_year.1") and ("facenumber_in_poster) as they have no use and contains many null values


```python
df.drop(["title_year.1","facenumber_in_poster"], axis = 1 , inplace = True)
df.head()
```

```python
df.shape
```

there is many approaches we can take regarding the null values, it depends on what we will do for this data, for example we could use it for a machine learning model so we can fill it with the mean or other in some columns, in this case we will just drop the nulls 

```python
# drop the null values
df = df.dropna()
```

```python
df.shape
```

```python
# checking for duplicates 
df.duplicated().any()
```

```python
# explore the "director_facebook_likes" column
df["director_facebook_likes"].unique()
```

an issue needs to be solved 

the column has a value = "475" .. so we will change it just to 475

```python
df["director_facebook_likes"].str.contains("""""")
```

there is no other values contain ("")

```python
# replace ("475") with 475
df["director_facebook_likes"].replace('"475"', "475", inplace = True)
```

```python
df.head()
```

#### we need to clean the ("movie_title") column as it has unwanted characters

```python
df["movie_title"] = df["movie_title"].str.replace("?" , " ")
df["movie_title"]= df["movie_title"].str.replace("ÿ" , " ")
```

```python
df.head()
```

```python
df.info()
```

##### now we will change the data type for some columns like ("director_facebook_likes" , "num_voted_users" , "cast_total_facebook_likes" , "facenumber_in_poster " , "actor_2_facebook_likes")


```python
# columns that we need to change their data types
col = "director_facebook_likes" , "num_voted_users" , "cast_total_facebook_likes", "actor_2_facebook_likes"
```

```python
for x in col:
    df[x] = df[x].astype(int)
    
```

```python
df.info()
```

```python
df
```

```python
# we will reset our index
df.reset_index(inplace = True)
```

```python
# drop the old index
df.drop("index",axis = 1, inplace = True)
```

```python
df
```

```python
df.to_csv("movies_cleaned.csv")
```
